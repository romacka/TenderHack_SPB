```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
contracts = pd.read_csv('data/contracts.csv', sep=',')
```

```python
contracts
```

```python
def category(price):
    if price > 3e6:
        return '>3кк'
    if price > 6.5e5:
        return '>650к<3кк'
    if price > 5e5:
        return '>500к<650к'
    if price > 1e5:
        return '>100к<500к'
    else:
        return '<100к'
```

```python
contracts['price_category'] = contracts['price'].apply(category)
```

```python
contracts.to_csv('contracts_with_category.csv', sep=';')
```
